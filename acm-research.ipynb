```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#read dataset

filename = "/kaggle/input/carsforsale/cars_raw.csv"
df = pd.read_csv(filename)
df
```

```python
# sort dataset by consumer rating

df = df.sort_values(by = ['ConsumerRating'], ascending = True)
df
```

```python
# generate graphs based on attributes and consumer reviews

attributes = ["Year", "Make", "Used/New", "Drivetrain", "MinMPG", "MaxMPG", "FuelType", "Transmission"] 
for attr in attributes:
    plt.figure(figsize=(32,20))
    plt.title("Attribute:" + attr + " Consumer Reviews")
    uniq = pd.unique(df[attr].values.ravel())
    for attri in uniq:
        attrib = df.loc[df[attr] == attri][attr]
        rating = df.loc[df[attr] == attri]["ConsumerReviews"]
        plt.scatter(attrib, rating, label = attri)
    ax = plt.gca()
    # ax.set_yscale('log')
#     ax.set_xticklabels([])
    plt.ylabel('reviews')
    plt.legend()
```

```python
# generate graphs based on attributes and consumer rating

attributes = ["Year", "Make", "Used/New", "Drivetrain", "MinMPG", "MaxMPG", "FuelType", "Transmission", "Engine"] 
for attr in attributes:
    plt.figure(figsize=(32,20))
    plt.title("Attribute:" + attr + " Consumer Rating")
    uniq = pd.unique(df[attr].values.ravel())
    for attri in uniq:
        attrib = df.loc[df[attr] == attri][attr]
        rating = df.loc[df[attr] == attri]["ConsumerRating"]
        plt.scatter(attrib, rating, label = attri)
    ax = plt.gca()
    # ax.set_yscale('log')
#     ax.set_xticklabels([])
    plt.ylabel('Rating')
    plt.legend()
```

```python
# show most popular attributes in each state

attributes = ["Year", "Make", "Used/New", "Drivetrain", "MinMPG", "MaxMPG", "FuelType", "Transmission"] 
states = pd.unique(df["State"].values.ravel())
for state in states:
    for attr in attributes:
        uniq = df.loc[df["State"] == state][attr].value_counts()
        print("*****Most popular",attr, "in", state)
        print(uniq)
```
